```python
#  selecting the Area and Y2017 column
sel_2017_area=food_data.iloc[:, [1,10]]
print(sel_2017_area,'\n')

#grouping by area
gb=sel_2017_area.groupby('Area')['Y2017'].sum()
print(gb,'\n')

#sort the highest sum in 2017
highest_sum= gb.sort_values().iloc[-1]
print('The highest sum is:', highest_sum, '\n')

#sorting the 7th lowest sum in 2017
seventh_lowest_sum= gb.sort_values().iloc[6]
print('The 7th lowest sum is:', seventh_lowest_sum)
```

```python
# getting the total sum of wine produced in 2015 and  2018

wine_2015= food_data.groupby('Item')['Y2015'].sum()
wine_2018= food_data.groupby('Item')['Y2018'].sum()

print('The totatl sum of wine produced in 2015 is', wine_2015['Wine'])
print('The totatl sum of wine produced in 2018 is', wine_2018['Wine'])
```
